```python
#Dependencies and Setup
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')

#Files 
mouse = "data/mouse_drug_data.csv"
clinical = "data/clinicaltrial_data.csv"

mouse_df = pd.read_csv(mouse)
clinical_df = pd.read_csv(clinical)
```

```python
#Merges the dataframes and resets their indexs
df = clinical_df.merge(mouse_df, on="Mouse ID")
df = df.sort_values("Timepoint")
df = df.reset_index()
df = df.drop(columns = "index")
df.head()
```

```python
#Sorts the dataframes by drug and timepoint, with the only other column being their average tumor size
mean_tv_df = df.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean()
mean_tv_df = pd.DataFrame(mean_tv_df)
mean_tv_df.head()
```

```python
#Uses the .sem() or standard error function to create a dataframe that shows the standard error
set_df = df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem()
set_df = pd.DataFrame(set_df)
set_df.head()
```

```python
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.unstack.html?highlight=unstack#pandas.DataFrame.unstack
#Uses the "Unstack" function to re-format the dataframe and display it by Tumor Volume
reformat_df = mean_tv_df.unstack(level = 0)
reformat_df = reformat_df["Tumor Volume (mm3)"]
reformat_df
```

```python
#Sets the x-axis for our scatter plot
x_limit = 50
x_axis = np.arange(0, x_limit, 5)

#Adds our points onto the graph, Settings the legend before adding the lines so the legend doesnt repeat
plt.scatter(x_axis, reformat_df["Capomulin"], marker="v", facecolors="C0", edgecolors="black", label = "Capomulin")
plt.scatter(x_axis, reformat_df["Ceftamin"], marker="o", facecolors="C1", edgecolors="black", label = "Ceftamin")
plt.scatter(x_axis, reformat_df["Infubinol"], marker="1", facecolors="C2", edgecolors="black", label = "Infubinol")
plt.scatter(x_axis, reformat_df["Ketapril"], marker="*", facecolors="C3", edgecolors="black", label = "Ketapril")
plt.scatter(x_axis, reformat_df["Naftisol"], marker="P", facecolors="C4", edgecolors="black", label = "Naftisol")
plt.scatter(x_axis, reformat_df["Placebo"], marker="8", facecolors="C5", edgecolors="black", label = "Placebo")
plt.scatter(x_axis, reformat_df["Propriva"], marker="s", facecolors="C6", edgecolors="black", label = "Propriva")
plt.scatter(x_axis, reformat_df["Ramicane"], marker="X", facecolors="C7", edgecolors="black", label = "Ramicane")
plt.scatter(x_axis, reformat_df["Stelasyn"], marker="D", facecolors="C8", edgecolors="black", label = "Stelasyn")
plt.scatter(x_axis, reformat_df["Zoniferol"], marker="4", facecolors="C9", edgecolors="black", label = "Zoniferol")

plt.legend(loc="best")

plt.plot(x_axis, reformat_df["Capomulin"], color="C0")
plt.plot(x_axis, reformat_df["Ceftamin"], color="C1")
plt.plot(x_axis, reformat_df["Infubinol"], color="C2")
plt.plot(x_axis, reformat_df["Ketapril"], color="C3")
plt.plot(x_axis, reformat_df["Naftisol"], color="C4")
plt.plot(x_axis, reformat_df["Placebo"], color="C5")
plt.plot(x_axis, reformat_df["Propriva"], color="C6")
plt.plot(x_axis, reformat_df["Ramicane"], color="C7")
plt.plot(x_axis, reformat_df["Stelasyn"], color="C8")
plt.plot(x_axis, reformat_df["Zoniferol"], color="C9")

#Sets the labels
plt.title("Drug Effectiveness", size = 20)
plt.ylabel("Tumor Size", size = 18)
plt.xlabel("Time", size = 18)

#Sets a nice ole grid
plt.grid(linestyle = "dashed")

#Sets the size of the graph
plt.rcParams["figure.figsize"] = [10,8]
```

```python
#Creates a DataFrame with mice who survived by the timepoint and drugs
survival_df = df.groupby(["Drug", "Timepoint"])["Mouse ID"].count()
survival_df = pd.DataFrame(survival_df)
survival_df = survival_df.rename(columns={"Mouse ID": "Mouse Count"})
survival_df.head()
```

```python
#Uses the "Unstack" function to re-format the dataframe and display it by Mouse Count
reformat_survival_df = survival_df.unstack(0)
reformat_survival_df = reformat_survival_df["Mouse Count"]
reformat_survival_df.head()
```
